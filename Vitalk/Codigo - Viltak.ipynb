```python
def bolha(rec,reset):
    copia = rec
    ind = reset
    #print('rec bolha',copia)
    #print("BOLHA",ind_nos)
    for i in range(0,len(copia)):
        for j in range(i,len(copia)):
            if (copia[j]>=copia[i]):
                copia[j],copia[i] = copia[i],copia[j]
                ind[j],ind[i] = ind[i],ind[j]
   # print("BOLHA",ind_nos)
     
```

```python
def converte_float(string):
    strin = []
    for x in string:
        if(x==','):
            break
        if(x!='.'):
            strin.append(x)
        
    convert = ''
    for x in strin:
        convert+=x
    return float(convert)
```

```python
converte_float('1.200,50')
```

```python
datas = ['2021-0'+str(x) for x in range(1,10)]
datas.append('2021-10')
datas.append('2021-11')
datas.append('2021-12')
negocios_por_mes = []
cont = 0

for i in datas:
    cont = 0
    for x in negocios_fechados.index:
        if(i in negocios_fechados['Negócio - Data de ganho'][x]):
            cont+=converte_float(negocios_fechados['Negócio - Valor'][x])
            
    negocios_por_mes.append(cont)

meses = ['jan', 'fev', 'mar', 'abr', 'mai', 'jun', 'jul', 'ago', 'set', 'out', 'nov', 'dez'] 
for x in range(len(meses)):
    print('Valor Total Vendido Mensalmente no mês de '+meses[x]+': '+str(negocios_por_mes[x]))
plt.figure(figsize=(20,3))
plt.subplot(121)
plt.bar(meses,negocios_por_mes)
plt.title('Valor Total Vendido Mensalmente')
plt.xlabel("Meses")
plt.ylabel('Valor Total')
plt.subplot(122)
plt.title('Valor Total Vendido Mensalmente')
plt.xlabel("Meses")
plt.ylabel('Valor Total')
plt.plot(meses,negocios_por_mes)
plt.show()
```
